```python
# Q9: Forecasting Electricity Access to 2030
from sklearn.model_selection import train_test_split

def forecast_country_access(country_name, access_type='Rural'):
    """Forecast electricity access for a specific country"""
    country_data = df[df['country'] == country_name].copy()
    country_data = country_data.dropna(subset=[f'Electricity_Access_{access_type}_Percentage'])
    
    if len(country_data) < 5:
        return None, None, None
    
    # Prepare features
    X = country_data[['year']].values
    y = country_data[f'Electricity_Access_{access_type}_Percentage'].values
    
    # Train model
    model = LinearRegression()
    model.fit(X, y)
    
    # Forecast to 2030
    future_years = np.arange(2017, 2031).reshape(-1, 1)
    forecast = model.predict(future_years)
    forecast = np.clip(forecast, 0, 100)  # Clip to 0-100%
    
    return future_years.flatten(), forecast, model.score(X, y)

# Forecast for select countries
countries_to_forecast = ['India', 'Nigeria', 'Bangladesh', 'Ethiopia', 'Kenya']
fig = make_subplots(rows=3, cols=2, subplot_titles=countries_to_forecast[:6],
                    vertical_spacing=0.1, horizontal_spacing=0.15)

for idx, country in enumerate(countries_to_forecast):
    row = (idx // 2) + 1
    col = (idx % 2) + 1
    
    # Historical data
    hist_data = df[df['country'] == country]
    
    # Forecast
    years, forecast, r2 = forecast_country_access(country, 'Rural')
    
    if forecast is not None:
        # Historical trace
        fig.add_trace(
            go.Scatter(x=hist_data['year'], 
                      y=hist_data['Electricity_Access_Rural_Percentage'],
                      mode='markers+lines',
                      name=f'{country} (Historical)',
                      line=dict(width=2),
                      showlegend=(idx==0)),
            row=row, col=col
        )
        
        # Forecast trace
        fig.add_trace(
            go.Scatter(x=years, 
                      y=forecast,
                      mode='lines',
                      name=f'{country} (Forecast)',
                      line=dict(dash='dash', width=2),
                      showlegend=(idx==0)),
            row=row, col=col
        )

fig.update_layout(height=900, title_text="🔮 Rural Electricity Access Forecast to 2030")
fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Rural Access (%)")
fig.show()
```
