## 1.📦 Import required libraries

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set styles
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: f'{x:,.2f}')

print("✅ Libraries imported successfully!")
```

## 2. 📂Load dataset

```python
# Load dataset
df = pd.read_csv('global_electricity_access_economic_indicators.csv')
print(f"📊 Dataset Shape: {df.shape}")
print(f"📅 Date Range: {df['date'].min()} to {df['date'].max()}")
print(f"🌍 Number of Countries: {df['country'].nunique()}")
print("\n🔍 First 5 rows:")
display(df.head())
```

```python
# Data Overview
print("📋 DATA OVERVIEW")
print("="*50)
print(f"Total Records: {len(df):,}")
print(f"Total Countries/Regions: {df['country'].nunique()}")
print(f"Time Period: {df['date'].min()} - {df['date'].max()}")
print(f"Years Covered: {df['date'].max() - df['date'].min() + 1} years")
print("\n📊 Column Information:")
print("="*50)

# Check data types and missing values
info_df = pd.DataFrame({
    'Data Type': df.dtypes,
    'Missing Values': df.isnull().sum(),
    'Missing %': (df.isnull().sum() / len(df) * 100).round(2)
})
display(info_df)

print("\n📈 Basic Statistics:")
print("="*50)
display(df.describe())
```

## 3. 🧹 Handle missing values

```python
# Handle missing values
print("🧹 DATA CLEANING")
print("="*50)

# Store original shape
original_shape = df.shape

# Create a copy for cleaning
df_clean = df.copy()

# For time series analysis, we'll handle missing values appropriately
# For electricity access columns, forward fill for each country
electricity_cols = ['Electricity_Access_Urban_Percentage', 
                    'Electricity_Access_Rural_Percentage',
                    'Total_Electricity_Output_GWh']

for col in electricity_cols:
    df_clean[col] = df_clean.groupby('country')[col].transform(
        lambda x: x.ffill().bfill()
    )

# Drop any remaining rows with critical missing values
df_clean = df_clean.dropna(subset=['GDP_Per_Capita_Current_USD', 'Total_Population'])

print(f"Records before cleaning: {original_shape[0]:,}")
print(f"Records after cleaning: {len(df_clean):,}")
print(f"Records removed: {original_shape[0] - len(df_clean):,}")
print(f"Percentage retained: {(len(df_clean)/original_shape[0]*100):.1f}%")
```

## 4. 🟢 Q1 —Countries with 100% electricity access by 2015

```python
# 1. Countries with 100% electricity access by 2015
print("🔋 COUNTRIES WITH 100% ELECTRICITY ACCESS BY 2015")
print("="*60)

# Filter data for 2015
df_2015 = df_clean[df_clean['date'] == 2015].copy()

# Calculate overall access (average of urban and rural)
df_2015['Overall_Electricity_Access'] = (
    df_2015['Electricity_Access_Urban_Percentage'] + 
    df_2015['Electricity_Access_Rural_Percentage']
) / 2

# Get countries with 100% access
full_access_countries = df_2015[
    (df_2015['Electricity_Access_Urban_Percentage'] >= 99.9) & 
    (df_2015['Electricity_Access_Rural_Percentage'] >= 99.9)
]['country'].unique()

print(f"Number of countries with 100% access: {len(full_access_countries)}")
print("\nCountries with 100% electricity access (both urban & rural):")
for i, country in enumerate(sorted(full_access_countries), 1):
    print(f"{i:2}. {country}")

# Visualization
fig = px.bar(df_2015.nlargest(20, 'Overall_Electricity_Access'),
             x='country', y='Overall_Electricity_Access',
             title='Top 20 Countries by Electricity Access (2015)',
             color='Overall_Electricity_Access',
             color_continuous_scale='viridis',
             labels={'Overall_Electricity_Access': 'Electricity Access (%)'})
fig.update_layout(xaxis_tickangle=45)
fig.show()
```

## 5. 🟢 Q2 —Global average rural electricity access trend (2000-2016)

```python
# 2. Global average rural electricity access trend (2000-2016)
print("📈 GLOBAL AVERAGE RURAL ELECTRICITY ACCESS TREND (2000-2016)")
print("="*60)

# Filter years and calculate global averages
global_trend = df_clean[df_clean['date'] <= 2016].groupby('date').agg({
    'Electricity_Access_Rural_Percentage': 'mean',
    'Electricity_Access_Urban_Percentage': 'mean'
}).reset_index()

# Calculate overall access
global_trend['Overall_Access'] = (
    global_trend['Electricity_Access_Urban_Percentage'] * 0.4 + 
    global_trend['Electricity_Access_Rural_Percentage'] * 0.6
)

print(f"Rural Access in 2000: {global_trend.iloc[0]['Electricity_Access_Rural_Percentage']:.1f}%")
print(f"Rural Access in 2016: {global_trend.iloc[-1]['Electricity_Access_Rural_Percentage']:.1f}%")
print(f"Improvement: {global_trend.iloc[-1]['Electricity_Access_Rural_Percentage'] - global_trend.iloc[0]['Electricity_Access_Rural_Percentage']:.1f}%")

# Visualization
fig = make_subplots(rows=2, cols=1, 
                   subplot_titles=('Global Electricity Access Trends', 'Access Gap Over Time'),
                   vertical_spacing=0.15)

# Line chart for access trends
fig.add_trace(
    go.Scatter(x=global_trend['date'], y=global_trend['Electricity_Access_Urban_Percentage'],
               mode='lines+markers', name='Urban Access', line=dict(color='#1f77b4', width=3)),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=global_trend['date'], y=global_trend['Electricity_Access_Rural_Percentage'],
               mode='lines+markers', name='Rural Access', line=dict(color='#ff7f0e', width=3)),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=global_trend['date'], y=global_trend['Overall_Access'],
               mode='lines', name='Overall Access', line=dict(color='#2ca02c', width=3, dash='dash')),
    row=1, col=1
)

# Area chart for access gap
global_trend['Access_Gap'] = (global_trend['Electricity_Access_Urban_Percentage'] - 
                             global_trend['Electricity_Access_Rural_Percentage'])

fig.add_trace(
    go.Scatter(x=global_trend['date'], y=global_trend['Access_Gap'],
               mode='lines', fill='tozeroy', name='Urban-Rural Gap',
               line=dict(color='#d62728', width=2), fillcolor='rgba(214, 39, 40, 0.3)'),
    row=2, col=1
)

fig.update_layout(height=800, showlegend=True, title_text="Global Electricity Access Analysis")
fig.update_xaxes(title_text="Year", row=2, col=1)
fig.update_yaxes(title_text="Access (%)", row=1, col=1)
fig.update_yaxes(title_text="Access Gap (%)", row=2, col=1)
fig.show()
```

## 6. 🟢 Q3—Top 10 countries by GDP per capita in 2023

```python
# 3. Top 10 countries by GDP per capita in 2023
print("💰 TOP 10 COUNTRIES BY GDP PER CAPITA (2023)")
print("="*60)

# Filter for 2023 and get top 10
df_2023 = df_clean[df_clean['date'] == 2023].copy()
top_gdp = df_2023.nlargest(10, 'GDP_Per_Capita_Current_USD')[['country', 'GDP_Per_Capita_Current_USD', 'Total_Population']]

print("Top 10 Countries by GDP per capita:")
display(top_gdp)

# Visualization - Horizontal Bar Chart
fig = px.bar(top_gdp.sort_values('GDP_Per_Capita_Current_USD'),
             y='country', x='GDP_Per_Capita_Current_USD',
             orientation='h',
             title='Top 10 Countries by GDP per Capita (2023)',
             color='GDP_Per_Capita_Current_USD',
             color_continuous_scale='sunset',
             hover_data=['Total_Population'],
             labels={'GDP_Per_Capita_Current_USD': 'GDP per Capita (USD)', 
                    'Total_Population': 'Population'})

fig.update_layout(yaxis={'categoryorder': 'total ascending'},
                  coloraxis_showscale=False)
fig.show()

# Additional: GDP Distribution
fig = px.box(df_2023, y='GDP_Per_Capita_Current_USD',
             title='Global GDP per Capita Distribution (2023)',
             labels={'GDP_Per_Capita_Current_USD': 'GDP per Capita (USD)'})
fig.show()
```

## 7. 🟢 Q4—Global population change (2000-2024)

```python
# 4. Global population change (2000-2024)
print("👥 GLOBAL POPULATION CHANGE (2000-2024)")
print("="*60)

# Calculate global population by year
global_population = df_clean.groupby('date')['Total_Population'].sum().reset_index()

# Calculate growth metrics
pop_2000 = global_population[global_population['date'] == 2000]['Total_Population'].values[0]
pop_2024 = global_population[global_population['date'] == 2024]['Total_Population'].values[0]
growth = pop_2024 - pop_2000
growth_pct = (growth / pop_2000) * 100

print(f"Global Population in 2000: {pop_2000:,.0f}")
print(f"Global Population in 2024: {pop_2024:,.0f}")
print(f"Total Growth: {growth:,.0f} people")
print(f"Percentage Growth: {growth_pct:.1f}%")
print(f"Average Annual Growth: {growth/24:,.0f} people")

# Visualization - Area Chart
fig = px.area(global_population, x='date', y='Total_Population',
              title='Global Population Growth (2000-2024)',
              labels={'Total_Population': 'Population', 'date': 'Year'},
              color_discrete_sequence=['#17becf'])

# Add annotations
fig.add_annotation(x=2000, y=pop_2000,
                   text=f"{pop_2000:,.0f}",
                   showarrow=True,
                   arrowhead=1)
fig.add_annotation(x=2024, y=pop_2024,
                   text=f"{pop_2024:,.0f}",
                   showarrow=True,
                   arrowhead=1)

fig.update_layout(hovermode='x unified')
fig.show()

# Population growth rate
global_population['Growth_Rate'] = global_population['Total_Population'].pct_change() * 100

fig = px.line(global_population, x='date', y='Growth_Rate',
              title='Global Population Growth Rate (%)',
              labels={'Growth_Rate': 'Annual Growth Rate (%)', 'date': 'Year'},
              markers=True)
fig.update_traces(line_color='#e377c2', line_width=3)
fig.add_hline(y=0, line_dash="dash", line_color="gray")
fig.show()
```

## 8. 🟡 Q5—Correlation between GDP per capita and electricity access

The Pearson correlation is defined as:
$$r = \frac{\sum(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum(x_i-\bar{x})^2 \cdot \sum(y_i-\bar{y})^2}}$$

```python
# 5. Correlation between GDP per capita and electricity access
print("📊 CORRELATION: GDP PER CAPITA VS ELECTRICITY ACCESS")
print("="*60)

# Prepare correlation data (using 2023 data)
corr_data = df_clean[df_clean['date'] == 2023].copy()
corr_data['Overall_Access'] = (
    corr_data['Electricity_Access_Urban_Percentage'] * 0.4 + 
    corr_data['Electricity_Access_Rural_Percentage'] * 0.6
)

# Calculate correlations
correlations = corr_data[['GDP_Per_Capita_Current_USD', 
                         'Electricity_Access_Urban_Percentage',
                         'Electricity_Access_Rural_Percentage',
                         'Overall_Access']].corr()

print("Correlation Matrix:")
display(correlations)

print(f"\nKey Correlations:")
print(f"GDP vs Urban Access: {correlations.loc['GDP_Per_Capita_Current_USD', 'Electricity_Access_Urban_Percentage']:.3f}")
print(f"GDP vs Rural Access: {correlations.loc['GDP_Per_Capita_Current_USD', 'Electricity_Access_Rural_Percentage']:.3f}")
print(f"GDP vs Overall Access: {correlations.loc['GDP_Per_Capita_Current_USD', 'Overall_Access']:.3f}")

# Visualization - Scatter Plots
fig = make_subplots(rows=1, cols=2,
                   subplot_titles=('GDP vs Urban Electricity Access', 
                                  'GDP vs Rural Electricity Access'),
                   horizontal_spacing=0.15)

# Log scale for GDP for better visualization
corr_data['Log_GDP'] = np.log10(corr_data['GDP_Per_Capita_Current_USD'] + 1)

# Urban access scatter
fig.add_trace(
    go.Scatter(x=corr_data['Log_GDP'], y=corr_data['Electricity_Access_Urban_Percentage'],
               mode='markers', name='Urban',
               marker=dict(size=8, color='#1f77b4', opacity=0.7,
                          line=dict(width=1, color='DarkSlateGrey')),
               text=corr_data['country'],
               hovertemplate='<b>%{text}</b><br>Log GDP: %{x:.2f}<br>Urban Access: %{y:.1f}%'),
    row=1, col=1
)

# Rural access scatter
fig.add_trace(
    go.Scatter(x=corr_data['Log_GDP'], y=corr_data['Electricity_Access_Rural_Percentage'],
               mode='markers', name='Rural',
               marker=dict(size=8, color='#ff7f0e', opacity=0.7,
                          line=dict(width=1, color='DarkSlateGrey')),
               text=corr_data['country'],
               hovertemplate='<b>%{text}</b><br>Log GDP: %{x:.2f}<br>Rural Access: %{y:.1f}%'),
    row=1, col=2
)

fig.update_layout(height=500, showlegend=False,
                  title_text="Relationship between GDP and Electricity Access (2023)")
fig.update_xaxes(title_text="Log10(GDP per Capita)", row=1, col=1)
fig.update_xaxes(title_text="Log10(GDP per Capita)", row=1, col=2)
fig.update_yaxes(title_text="Urban Access (%)", row=1, col=1)
fig.update_yaxes(title_text="Rural Access (%)", row=1, col=2)
fig.show()

# Heatmap
fig = px.imshow(correlations,
                text_auto=True,
                aspect="auto",
                color_continuous_scale='RdBu_r',
                title="Correlation Heatmap (2023)")
fig.update_layout(height=500)
fig.show()
```

## 9. 🟡 Q6—Regions with fastest rural electricity access growth

```python
# 6. Regions with fastest rural electricity access growth
print("🚀 FASTEST GROWING REGIONS IN RURAL ELECTRICITY ACCESS")
print("="*60)

# Define regions (simplified - in practice, use proper regional classification)
# For this example, we'll group by continents manually or use World Bank regions
# Let's create a simplified regional grouping based on country names
def assign_region(country):
    """Simple regional assignment based on country names"""
    if any(keyword in country.lower() for keyword in ['asia', 'china', 'india', 'japan', 'korea']):
        return 'Asia'
    elif any(keyword in country.lower() for keyword in ['africa', 'kenya', 'nigeria', 'south africa', 'egypt']):
        return 'Africa'
    elif any(keyword in country.lower() for keyword in ['europe', 'germany', 'france', 'italy', 'spain', 'uk']):
        return 'Europe'
    elif any(keyword in country.lower() for keyword in ['america', 'usa', 'canada', 'mexico', 'brazil', 'argentina']):
        return 'Americas'
    else:
        return 'Other'

df_clean['Region'] = df_clean['country'].apply(assign_region)

# Calculate growth rates by region
region_growth = []
for region in df_clean['Region'].unique():
    region_data = df_clean[df_clean['Region'] == region]
    if len(region_data[region_data['date'] == 2000]) > 0 and len(region_data[region_data['date'] == 2016]) > 0:
        start = region_data[region_data['date'] == 2000]['Electricity_Access_Rural_Percentage'].mean()
        end = region_data[region_data['date'] == 2016]['Electricity_Access_Rural_Percentage'].mean()
        if not np.isnan(start) and not np.isnan(end):
            growth = end - start
            region_growth.append({
                'Region': region,
                'Start_2000': start,
                'End_2016': end,
                'Growth': growth,
                'Growth_Percentage': ((end - start) / start * 100) if start > 0 else 0
            })

growth_df = pd.DataFrame(region_growth).sort_values('Growth_Percentage', ascending=False)

print("Regional Rural Electricity Access Growth (2000-2016):")
display(growth_df)

# Visualization
fig = px.bar(growth_df, x='Region', y='Growth',
             title='Rural Electricity Access Growth by Region (2000-2016)',
             color='Growth',
             color_continuous_scale='tealrose',
             labels={'Growth': 'Percentage Point Increase'})

fig.update_layout(xaxis_title="Region", yaxis_title="Growth (Percentage Points)")
fig.show()

# Line chart showing regional trends
regional_trends = df_clean.groupby(['date', 'Region'])['Electricity_Access_Rural_Percentage'].mean().reset_index()

fig = px.line(regional_trends, x='date', y='Electricity_Access_Rural_Percentage',
              color='Region', title='Rural Electricity Access Trends by Region',
              markers=True, line_shape='spline')
fig.update_layout(height=600)
fig.show()
```

## 10. 🟡 Q7—Electricity output vs population

$$E_{per\_capita} = \frac{\text{Total Electricity Output (GWh)} \times 10^6}{\text{Total Population}}$$

```python
# 7. Electricity output vs population
print("⚡ ELECTRICITY OUTPUT VS POPULATION ANALYSIS")
print("="*60)

# Prepare data for 2023
energy_pop = df_clean[df_clean['date'] == 2023].dropna(subset=['Total_Electricity_Output_GWh', 'Total_Population']).copy()

# Calculate per capita electricity
energy_pop['Electricity_per_Capita'] = energy_pop['Total_Electricity_Output_GWh'] * 1e6 / energy_pop['Total_Population']  # kWh per capita

print(f"Countries with data: {len(energy_pop)}")
print(f"Global average electricity per capita: {energy_pop['Electricity_per_Capita'].mean():.0f} kWh")

# Top and bottom countries
print("\nTop 10 Countries by Electricity per Capita:")
display(energy_pop.nlargest(10, 'Electricity_per_Capita')[['country', 'Electricity_per_Capita', 'Total_Population']])

print("\nBottom 10 Countries by Electricity per Capita:")
display(energy_pop.nsmallest(10, 'Electricity_per_Capita')[['country', 'Electricity_per_Capita', 'Total_Population']])

# Visualization - Bubble Chart
fig = px.scatter(energy_pop, x='Total_Population', y='Total_Electricity_Output_GWh',
                 size='Electricity_per_Capita', color='Electricity_per_Capita',
                 hover_name='country',
                 log_x=True, log_y=True,
                 size_max=60,
                 color_continuous_scale='viridis',
                 title='Electricity Output vs Population (2023) - Bubble Size = kWh per Capita',
                 labels={'Total_Population': 'Population (log scale)',
                        'Total_Electricity_Output_GWh': 'Electricity Output - GWh (log scale)',
                        'Electricity_per_Capita': 'kWh per Capita'})

fig.update_layout(height=700)
fig.show()

# Histogram of electricity per capita
fig = px.histogram(energy_pop, x='Electricity_per_Capita',
                   nbins=50,
                   title='Distribution of Electricity Consumption per Capita',
                   labels={'Electricity_per_Capita': 'kWh per Capita'},
                   color_discrete_sequence=['#17becf'])
fig.update_layout(bargap=0.05)
fig.show()
```

## 11. 🟡 Q8—Countries with largest urban-rural electricity gap

## ## 12. 🔴 Q9—Forecasting electricity access to 2030

Using the linear trend model:
$$\hat{y} = \beta_0 + \beta_1 t + \epsilon$$

```python
# 9. Forecasting electricity access to 2030
print("🔮 FORECASTING ELECTRICITY ACCESS TO 2030")
print("="*60)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# Prepare data for forecasting
forecast_data = df_clean.groupby('date').agg({
    'Electricity_Access_Urban_Percentage': 'mean',
    'Electricity_Access_Rural_Percentage': 'mean'
}).reset_index()

# Create features for forecasting
forecast_data['Year'] = forecast_data['date']
forecast_data['Year_Squared'] = forecast_data['Year'] ** 2

# Train models
models = {}
predictions = {}

for access_type in ['Urban', 'Rural']:
    col = f'Electricity_Access_{access_type}_Percentage'
    
    # Prepare training data
    X = forecast_data[['Year', 'Year_Squared']]
    y = forecast_data[col]
    
    # Train model
    model = LinearRegression()
    model.fit(X, y)
    models[access_type] = model
    
    # Make predictions for 2025-2030
    future_years = pd.DataFrame({
        'Year': range(2025, 2031),
        'Year_Squared': [y**2 for y in range(2025, 2031)]
    })
    
    pred = model.predict(future_years)
    predictions[access_type] = pred
    
    # Calculate metrics
    y_pred_train = model.predict(X)
    mae = mean_absolute_error(y, y_pred_train)
    r2 = r2_score(y, y_pred_train)
    
    print(f"\n{access_type} Access Model:")
    print(f"  R² Score: {r2:.3f}")
    print(f"  MAE: {mae:.2f}%")
    print(f"  Coefficients: Year={model.coef_[0]:.3f}, Year²={model.coef_[1]:.4f}")
    print(f"  Intercept: {model.intercept_:.2f}")

# Create forecast dataframe
forecast_df = pd.DataFrame({
    'Year': range(2025, 2031),
    'Urban_Forecast': predictions['Urban'],
    'Rural_Forecast': predictions['Rural']
})

print("\n📈 Forecasted Electricity Access Rates:")
display(forecast_df)

# Visualization
fig = go.Figure()

# Historical data
fig.add_trace(go.Scatter(x=forecast_data['date'], 
                         y=forecast_data['Electricity_Access_Urban_Percentage'],
                         mode='lines+markers',
                         name='Urban (Historical)',
                         line=dict(color='#1f77b4', width=2)))

fig.add_trace(go.Scatter(x=forecast_data['date'], 
                         y=forecast_data['Electricity_Access_Rural_Percentage'],
                         mode='lines+markers',
                         name='Rural (Historical)',
                         line=dict(color='#ff7f0e', width=2)))

# Forecast
fig.add_trace(go.Scatter(x=forecast_df['Year'], 
                         y=forecast_df['Urban_Forecast'],
                         mode='lines+markers',
                         name='Urban (Forecast)',
                         line=dict(color='#1f77b4', width=2, dash='dash')))

fig.add_trace(go.Scatter(x=forecast_df['Year'], 
                         y=forecast_df['Rural_Forecast'],
                         mode='lines+markers',
                         name='Rural (Forecast)',
                         line=dict(color='#ff7f0e', width=2, dash='dash')))

# SDG-7 Target line (100% by 2030)
fig.add_hline(y=100, line_dash="dot", line_color="green", 
              annotation_text="SDG-7 Target: 100% by 2030")

fig.update_layout(title='Electricity Access Forecast to 2030 (SDG-7)',
                  xaxis_title="Year",
                  yaxis_title="Electricity Access (%)",
                  height=600)
fig.show()

# Check SDG-7 progress
print("\n🎯 SDG-7 PROGRESS CHECK:")
print("="*40)

urban_target = forecast_df[forecast_df['Year'] == 2030]['Urban_Forecast'].values[0]
rural_target = forecast_df[forecast_df['Year'] == 2030]['Rural_Forecast'].values[0]
overall_target = (urban_target * 0.4 + rural_target * 0.6)  # Weighted average

print(f"Projected Urban Access in 2030: {urban_target:.1f}%")
print(f"Projected Rural Access in 2030: {rural_target:.1f}%")
print(f"Projected Overall Access in 2030: {overall_target:.1f}%")
print(f"\nGap to SDG-7 Target (100%): {100 - overall_target:.1f}%")

if overall_target >= 95:
    print("✅ On track to achieve SDG-7")
elif overall_target >= 80:
    print("⚠️ Moderate progress - needs acceleration")
else:
    print("❌ Off track - requires significant intervention")
```

## 13. 🔴 Q10—Clustering countries by electrification & economic profile

```python
# 10. Clustering countries by electrification & economic profile
print("🏷️ CLUSTERING COUNTRIES BY ELECTRIFICATION & ECONOMIC PROFILE")
print("="*60)

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Prepare data for clustering (using 2023 data)
cluster_data = df_clean[df_clean['date'] == 2023].copy()
cluster_data = cluster_data.dropna(subset=['GDP_Per_Capita_Current_USD',
                                          'Electricity_Access_Urban_Percentage',
                                          'Electricity_Access_Rural_Percentage',
                                          'Total_Electricity_Output_GWh']).copy()

# Create features for clustering
cluster_data['Overall_Access'] = (
    cluster_data['Electricity_Access_Urban_Percentage'] * 0.4 + 
    cluster_data['Electricity_Access_Rural_Percentage'] * 0.6
)

cluster_data['Access_Gap'] = (
    cluster_data['Electricity_Access_Urban_Percentage'] - 
    cluster_data['Electricity_Access_Rural_Percentage']
)

# Select features
features = ['GDP_Per_Capita_Current_USD', 'Overall_Access', 
           'Access_Gap', 'Total_Electricity_Output_GWh']
X = cluster_data[features]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using elbow method
inertia = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Visualization - Elbow Method
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(k_range), y=inertia,
                         mode='lines+markers',
                         line=dict(color='royalblue', width=3),
                         marker=dict(size=8)))
fig.update_layout(title='Elbow Method for Optimal k',
                  xaxis_title="Number of Clusters (k)",
                  yaxis_title="Inertia",
                  height=500)
fig.show()

# Apply KMeans with selected k (from elbow plot)
optimal_k = 4  # Based on elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)

# Add cluster labels to data
cluster_data['Cluster'] = cluster_labels

# Analyze clusters
cluster_summary = cluster_data.groupby('Cluster').agg({
    'GDP_Per_Capita_Current_USD': 'mean',
    'Overall_Access': 'mean',
    'Access_Gap': 'mean',
    'Total_Electricity_Output_GWh': 'mean',
    'country': 'count'
}).round(2)

cluster_summary.columns = ['Avg_GDP', 'Avg_Access', 'Avg_Gap', 'Avg_Output', 'Count']
cluster_summary = cluster_summary.sort_values('Avg_GDP')

print("\n📊 CLUSTER PROFILES:")
print("="*40)
for cluster_id in sorted(cluster_summary.index):
    stats = cluster_summary.loc[cluster_id]
    print(f"\nCluster {cluster_id} (Count: {stats['Count']} countries):")
    print(f"  • Avg GDP per capita: ${stats['Avg_GDP']:,.0f}")
    print(f"  • Avg Electricity Access: {stats['Avg_Access']:.1f}%")
    print(f"  • Avg Urban-Rural Gap: {stats['Avg_Gap']:.1f}%")
    print(f"  • Avg Electricity Output: {stats['Avg_Output']:,.0f} GWh")

# Visualize clusters using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

cluster_data['PCA1'] = X_pca[:, 0]
cluster_data['PCA2'] = X_pca[:, 1]

# Scatter plot with clusters
fig = px.scatter(cluster_data, x='PCA1', y='PCA2',
                 color='Cluster',
                 hover_name='country',
                 hover_data=['GDP_Per_Capita_Current_USD', 'Overall_Access'],
                 title='Country Clusters by Electrification & Economic Profile',
                 labels={'PCA1': 'Principal Component 1',
                        'PCA2': 'Principal Component 2',
                        'Cluster': 'Cluster'},
                 color_continuous_scale='viridis')

fig.update_layout(height=700)
fig.show()

# Show sample countries from each cluster
print("\n🔍 SAMPLE COUNTRIES FROM EACH CLUSTER:")
print("="*40)
for cluster_id in sorted(cluster_data['Cluster'].unique()):
    cluster_countries = cluster_data[cluster_data['Cluster'] == cluster_id]['country'].tolist()
    print(f"\nCluster {cluster_id}:")
    print(f"  Sample: {', '.join(cluster_countries[:5])}")
    if len(cluster_countries) > 5:
        print(f"  (+ {len(cluster_countries) - 5} more countries)")
```

## ## 14. 🔴 Q11—Countries off-track for SDG-7 universal access

For each country, the required annual growth rate to reach 100% access by 2030:
$$r_{required} = \frac{100 - \text{Access}_{2016}}{2030 - 2016} = \frac{100 - \text{Access}_{2016}}{14}$$

```python
# 11. Countries off-track for SDG-7 universal access
print("⚠️ COUNTRIES OFF-TRACK FOR SDG-7 UNIVERSAL ACCESS")
print("="*60)

# Calculate current progress and required growth rate
latest_year = 2023
target_year = 2030
years_remaining = target_year - latest_year

# Get latest data
sdg_data = df_clean[df_clean['date'] == latest_year].copy()
sdg_data['Overall_Access'] = (
    sdg_data['Electricity_Access_Urban_Percentage'] * 0.4 + 
    sdg_data['Electricity_Access_Rural_Percentage'] * 0.6
)

# Calculate required annual growth rate to reach 100% by 2030
sdg_data['Required_Annual_Growth'] = (
    (100 - sdg_data['Overall_Access']) / years_remaining
)

# Calculate historical annual growth (2015-2023)
def calculate_historical_growth(country_df):
    country_df = country_df.copy()
    country_df['Overall_Access'] = (
        country_df['Electricity_Access_Urban_Percentage'] * 0.4 + 
        country_df['Electricity_Access_Rural_Percentage'] * 0.6
    )
    if len(country_df) >= 2:
        years = sorted(country_df['date'].unique())
        if 2015 in years and latest_year in years:
            start = country_df[country_df['date'] == 2015]['Overall_Access'].values[0]
            end = country_df[country_df['date'] == latest_year]['Overall_Access'].values[0]
            if not np.isnan(start) and not np.isnan(end) and start > 0:
                return (end - start) / (latest_year - 2015)
    return np.nan

historical_growth = []
for country in sdg_data['country'].unique():
    country_data = df_clean[df_clean['country'] == country]
    growth = calculate_historical_growth(country_data)
    historical_growth.append({'country': country, 'Historical_Growth': growth})

growth_df = pd.DataFrame(historical_growth)
sdg_data = sdg_data.merge(growth_df, on='country', how='left')

# Identify off-track countries
sdg_data['Growth_Gap'] = sdg_data['Required_Annual_Growth'] - sdg_data['Historical_Growth']
sdg_data['On_Track'] = sdg_data['Growth_Gap'] <= 0  # Negative or zero gap means on track

# Analysis
total_countries = len(sdg_data)
on_track = sdg_data['On_Track'].sum()
off_track = total_countries - on_track

print(f"📊 SDG-7 PROGRESS SNAPSHOT:")
print(f"   Total countries analyzed: {total_countries}")
print(f"   Countries on track: {on_track} ({on_track/total_countries*100:.1f}%)")
print(f"   Countries off track: {off_track} ({off_track/total_countries*100:.1f}%)")

# Most critical countries (largest growth gap)
critical_countries = sdg_data.nlargest(15, 'Growth_Gap')[['country', 
                                                         'Overall_Access',
                                                         'Historical_Growth',
                                                         'Required_Annual_Growth',
                                                         'Growth_Gap']]

print("\n🔴 TOP 15 COUNTRIES MOST OFF-TRACK:")
display(critical_countries)

# Visualization
fig = make_subplots(rows=2, cols=1,
                   subplot_titles=('Current Access vs Required Growth Rate',
                                  'Countries with Largest Growth Gap'),
                   vertical_spacing=0.2)

# Scatter: Current access vs required growth
fig.add_trace(
    go.Scatter(x=sdg_data['Overall_Access'], 
               y=sdg_data['Required_Annual_Growth'],
               mode='markers',
               marker=dict(size=8,
                          color=sdg_data['On_Track'].map({True: 'green', False: 'red'}),
                          opacity=0.7),
               text=sdg_data['country'],
               hovertemplate='<b>%{text}</b><br>Current: %{x:.1f}%<br>Required Growth: %{y:.2f}%',
               name=''),
    row=1, col=1
)

# Add reference lines
fig.add_hline(y=0, line_dash="dash", line_color="gray", row=1, col=1)
fig.add_vline(x=100, line_dash="dash", line_color="green", row=1, col=1)

# Bar chart: Largest growth gaps
fig.add_trace(
    go.Bar(x=critical_countries['country'],
           y=critical_countries['Growth_Gap'],
           marker_color='crimson',
           text=critical_countries['Growth_Gap'].round(2),
           textposition='outside'),
    row=2, col=1
)

fig.update_layout(height=900, showlegend=False)
fig.update_xaxes(title_text="Current Electricity Access (%)", row=1, col=1)
fig.update_yaxes(title_text="Required Annual Growth (%)", row=1, col=1)
fig.update_xaxes(title_text="Country", tickangle=45, row=2, col=1)
fig.update_yaxes(title_text="Additional Growth Needed (%)", row=2, col=1)
fig.show()

# World map showing on-track vs off-track
fig = px.choropleth(sdg_data,
                    locations='country',
                    locationmode='country names',
                    color='On_Track',
                    hover_name='country',
                    hover_data=['Overall_Access', 'Required_Annual_Growth', 'Historical_Growth'],
                    color_discrete_map={True: 'green', False: 'red'},
                    title='SDG-7 Electricity Access Progress (Green = On Track, Red = Off Track)',
                    labels={'On_Track': 'On Track for 2030'})

fig.update_layout(height=600)
fig.show()
```
